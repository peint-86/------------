```python
import numpy as np

def lec_word_tfidf_matrix(courses, word_list):
    """
    Create a TF-IDF matrix for lecture-word pairs.

    Args:
        courses (list of lists): List of lectures, where each lecture is a list of words.
        word_list (list): List of unique words.

    Returns:
        np.ndarray: TF-IDF matrix.
    """
    # Initialize an empty matrix
    tfidf_matrix = np.zeros((len(courses), len(word_list)))

    # Compute the document frequency (DF) for each word
    df = np.zeros(len(word_list))
    for i, word in enumerate(word_list):
        for course in courses:
            if word in course:
                df[i] += 1

    # Compute the TF-IDF values
    for i, course in enumerate(courses):
        for j, word in enumerate(word_list):
            # Compute term frequency (TF)
            tf = course.count(word) / len(course)
            # Compute inverse document frequency (IDF)
            idf = np.log(len(courses) / df[j])
            # Compute TF-IDF
            tfidf_matrix[i, j] = tf * idf

    return tfidf_matrix

# Example usage
courses = [
    ["machine", "learning", "deep", "learning"],
    ["natural", "language", "processing"],
    ["computer", "vision", "deep", "learning"],
]

word_list = ["machine", "learning", "deep", "natural", "language", "processing", "computer", "vision"]

# Call the function to create the TF-IDF matrix
tfidf_matrix = lec_word_tfidf_matrix(courses, word_list)

# Print the TF-IDF matrix
print(tfidf_matrix)
```
